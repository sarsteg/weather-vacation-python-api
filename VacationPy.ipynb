### Step 1: Create a map that displays a point for every city in the `city_data_df` DataFrame. The size of the point should be the humidity in each city.

```python
from IPython.display import IFrame

# Create a map centered at a specific location (e.g., the coordinates of the first city)
m = folium.Map(location=[city_data_df['Lat'].iloc[0], city_data_df['Lng'].iloc[0]], zoom_start=5)

# Add markers for each city
for index, row in city_data_df.iterrows():
    folium.CircleMarker(
        location=[row['Lat'], row['Lng']],
        radius=row['Humidity'],  # Use humidity value as marker size
        popup=row['City'],      # Display city name in the popup
        color='blue',           # Marker color
        fill=True,
        fill_color='blue'
    ).add_to(m)

# Save the map to an HTML file
m.save('city_humidity_map.html')

# Display the map in the Jupyter Notebook using an IFrame
# IFrame('city_humidity_map.html', width=700, height=450)
m
```

### Step 2: Narrow down the `city_data_df` DataFrame to find your ideal weather condition

```python
# Narrow down cities that fit criteria and drop any results with null values

# I want to vacation in a place that's...
    # warm, but not hot
    # just the right amount of humidity to avoid fluffy hair and bloody noses
    # not too windy
    # I don't want to use a passport to get there

# Create a new DataFrame fitting weather criteria

# Copy the data from the city_data_df DataFrame into a new DataFrame
sarahs_destinations_df = city_data_df.copy()

# Filter for temperatures between 70 and 90 degrees Fahrenheit (21-32 degrees Celsius)
sarahs_destinations_df = sarahs_destinations_df.loc[(sarahs_destinations_df['Max Temp'] >= 21) & (sarahs_destinations_df['Max Temp'] <= 32)]

# Filter for wind speeds less than 10 mph
sarahs_destinations_df = sarahs_destinations_df.loc[sarahs_destinations_df['Wind Speed'] < 10]

# Filter for cities that are not too humid or too dry
sarahs_destinations_df = sarahs_destinations_df.loc[(sarahs_destinations_df['Humidity'] <= 70) & (sarahs_destinations_df['Humidity'] >= 20)]

# Filter for locations in the US only
sarahs_destinations_df = sarahs_destinations_df.loc[sarahs_destinations_df['Country'] == 'US']

sarahs_destinations_df
```

```python
# Fix the index
sarahs_destinations_df.reset_index(drop=True, inplace=True)

# Display the DataFrame
sarahs_destinations_df
```

### Step 3: Create a new DataFrame called `hotel_df`.

```python
# Use the Pandas copy function to create DataFrame called hotel_df to store the city, country, coordinates, and humidity
hotel_df = sarahs_destinations_df[['City', 'Country', 'Lat', 'Lng', 'Humidity']].copy()

# Add an empty column, "Hotel Name," to the DataFrame so you can store the hotel found using the Geoapify API
hotel_df['Hotel Name'] = ''

# Display sample data
hotel_df.head()
```

### Step 4: For each city, use the Geoapify API to find the first hotel located within 10,000 metres of your coordinates.

```python
# Set parameters to search for a hotel
radius = 100000 #extra large for troubleshooting
params = {
    "categories":"accommodation.hotel",
    "apiKey": geoapify_key,
    "radius": radius,
}

# Print a message to follow up the hotel search
print("Starting hotel search")

# Iterate through the hotel_df DataFrame
for index, row in hotel_df.iterrows():
    # get latitude, longitude from the DataFrame
    latitude = row["Lat"]
    longitude = row["Lng"]
    
    # Add filter and bias parameters with the current city's latitude and longitude to the params dictionary
    # params["filter"] = f"category=accommodation&lat={latitude}&lon={longitude}"
    params["filter"] = f"circle:{longitude},{latitude},{radius}"
    params["bias"] = f"proximity:{longitude},{latitude}"
    # params["bias"] = "proximity"
    
    # Set base URL
    base_url = "https://api.geoapify.com/v2/places"


    # Make and API request using the params dictionaty & Convert to JSON
    name_address = requests.get(base_url, params=params).json()
    print(name_address)
    
    # Grab the first hotel from the results and store the name in the hotel_df DataFrame
    try:
        hotel_df.loc[index, "Hotel Name"] = name_address["features"][0]["properties"]["name"]
    # except (KeyError, IndexError):
    #     # If no hotel is found, set the hotel name as "No hotel found".
    #     hotel_df.loc[index, "Hotel Name"] = "No hotel found"
    except Exception as E:
        print(E)
        
    # Log the search results
    print(f"{hotel_df.loc[index, 'City']} - nearest hotel: {hotel_df.loc[index, 'Hotel Name']}")

# Display sample data
hotel_df
```
